###Interpolation Using Radial Basis Function

RBF is better than inverse distance weighting (IDW) because IDW uses observed values as local minima and maxima.

```python
from scipy.interpolate import Rbf
# 2-d tests - setup scattered data
x = np.random.rand(100)*4.0-2.0
y = np.random.rand(100)*4.0-2.0
z = x*np.exp(-x**2-y**2)
ti = np.linspace(-2.0, 2.0, 100)
XI, YI = np.meshgrid(ti, ti)
# use RBF
rbf = Rbf(x, y, z)
ZI = rbf(XI, YI)
# plot the result
plt.pcolor(XI, YI, ZI, cmap=cm.jet)
plt.scatter(x, y, 100, z, cmap=cm.jet)
plt.title('RBF interpolation - multiquadrics')
plt.axis('equal')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.colorbar();
```

###Data Preparation

```python
import pymysql as db
db.install_as_MySQLdb()

con = db.connect('mysql.guaminsects.net',
                 'aubreymoore',
                 'canada12',
                 'oryctes')
#cur = con.cursor(db.cursors.DictCursor)
cur = con.cursor()
cur.execute("DROP TABLE IF EXISTS tempTrapCatch")
cur.execute("""
CREATE TABLE tempTrapCatch 
SELECT trap_visit.trap_id AS trapID, latitude, longitude, visit_date as date2,
(
SELECT visit_date
FROM trap_visit
WHERE trap_id = trapID
AND visit_date < date2
ORDER BY visit_date DESC
LIMIT 1
) AS date1,
male_count + female_count + unsexed_count AS trapCatch
FROM trap_visit, trap
WHERE trap_visit.trap_id = trap.trap_id""")
cur.execute("""
UPDATE tempTrapCatch, trap 
SET date1 = date_deployed
WHERE trapID = trap_id
  AND date1 IS NULL""")
cur.execute("""
DELETE FROM tempTrapCatch
WHERE date1 IS NULL
  OR date2 IS NULL
  OR date1 = 0
  OR date2 = 2
  OR date1 >= date2""")
cur.execute("""
SELECT * FROM tempTrapCatch""")

rows = cur.fetchall()
cur.rowcount
```

```python
import csv
with open('trapData.csv', 'wb') as result:
    writer = csv.writer(result)
    writer.writerow(('trapId','latitude','longitude','date2','date1','trapCatch'))
    writer.writerows(rows)
'Data written to trapData.csv.'
```

```python
def daily(df):
# returns a daily data frame    
# Date is saved as a string because pandasql cannot handle the datetime64 data type
    nrows = df.shape[0]
    rows_list = []
    for r in range(nrows):
        row = df.loc[r,]
        trapdays = (row.date2-row.date1).days
        traprate = 1.0*row.trapCatch/trapdays
        for n in range(trapdays):
            dict1 = {}
            date = row.date1 + n*datetime.timedelta(1)
            dict1['trapId'] = row.trapId
            dict1['latitude'] = row.latitude
            dict1['longitude'] = row.longitude
            dict1['date'] = date.strftime('%Y%m%d')
            dict1['traprate'] = traprate
            rows_list.append(dict1)
    dailydf = pd.DataFrame(rows_list, columns = ['trapId','latitude','longitude','date','traprate'])   
    return dailydf

#dailydf = daily(smalldata)
#dailydf.head()
```

```python
import pyproj

# Convert latitude and longitude in degrees to UTM coordinates (meters)
def deg_to_utm(df):
    p = pyproj.Proj(proj='utm', zone=55, ellps='WGS84')
    df['longitude'],df['latitude'] = p(df['longitude'].values, df['latitude'].values)
    return df        
```

```python
import pandasql as sqldf

sql = """
SELECT trapId, latitude, longitude, AVG(traprate)
FROM dailydf
WHERE date >= '20150101' AND date <= '20150331'
GROUP BY trapId;"""
m = sqldf.sqldf(sql, locals())
m.head()
```

```python

#figsize(10,10)

xmin = data['longitude'].min() - 0.01
xmax = data['longitude'].max() + 0.01
ymin = data['latitude'].min() - 0.01
ymax = data['latitude'].max() + 0.01
x = m['longitude']
y = m['latitude']
z = m['AVG(traprate)']

xi = np.linspace(xmin, xmax, 200)
yi = np.linspace(ymin, ymax, 200)
XI, YI = np.meshgrid(xi, yi)
# use RBF
rbf = Rbf(x, y, z, function = 'inverse', smooth=-0.05)
ZI = rbf(XI, YI)
# plot the result
plt.pcolor(XI, YI, ZI, cmap=cm.jet)
plt.scatter(x, y, 30, z, cmap=cm.jet)
plt.title('RBF interpolation - multiquadrics')
#plt.axis('equal')
#plt.xlim(-2, 2)
#plt.ylim(-2, 2)
plt.colorbar();
```
